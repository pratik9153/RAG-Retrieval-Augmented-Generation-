## Data Parsing CSV 

```python
import pandas as pd 
import os
```

```python
from langchain_community.document_loaders import CSVLoader
```

```python
try:
    csv_loader = CSVLoader("data/csv/students.csv")
    docs = csv_loader.load()

    print(f"Loaded: {len(docs)} documents\n")

    for i, doc in enumerate(docs, start=1):

        # Print document number
        print(f"Document {i}")

        # Print separator
        print("-" * 40)

        # Print content preview
        print(f"Content Preview:\n{doc.page_content[:100]}")

        # Print metadata
        print(f"Metadata: {doc.metadata}")

        print()

except Exception as e:
    print(f"Error: {e}")
```

# Excle data

```python
from langchain_community.document_loaders import UnstructuredExcelLoader

print("UnstructuredExcelLoader")

try:
    excel_loader = UnstructuredExcelLoader("data/csv/dt.xlsx")
    docs = excel_loader.load()

    print(f"Loaded: {len(docs)} documents\n")

    for i, doc in enumerate(docs, start=1):

        print(f"Document {i}")
        print("-" * 40)
        print(f"Content Preview:\n{doc.page_content[:100]}")
        print(f"Metadata: {doc.metadata}")
        print()

except Exception as e:
    print(f"Error: {e}")
```

```python
from langchain_community.document_loaders import JSONLoader

print("JSONLoader")

try:
    json_loader = JSONLoader(
        file_path="data/json/students.json",
        jq_schema=".[]",          # Read each object in the array
        text_content=False
    )

    docs = json_loader.load()

    print(f"Loaded: {len(docs)} documents\n")

    for i, doc in enumerate(docs, start=1):

        # Print document number
        print(f"Document {i}")

        # Print separator
        print("-" * 40)

        # Print content preview
        print(f"Content Preview:\n{doc.page_content[:100]}")

        # Print metadata
        print(f"Metadata: {doc.metadata}")

        print()

except Exception as e:
    print(f"Error: {e}")
```
