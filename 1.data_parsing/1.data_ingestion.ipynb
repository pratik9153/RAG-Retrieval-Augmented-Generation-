## DirectoryLoader - Suitable for Multiple text files

```python
from langchain_community.document_loaders import DirectoryLoader

#load all the files from directory 

dir_loader = DirectoryLoader(
    "data/text_files/",
    glob="**/*",#pattern to match files
    loader_cls= TextLoader, #loader class to use
    loader_kwargs={'encoding':'utf-8'},
    show_progress=True
)


d1 = dir_loader.load()


print(f"Loaded {len(d1)} documment")

for i , doc in enumerate(d1):
    print(f"\nDocument {i+1}:")
    print(f" Source : {doc.metadata['source']}")
    print(f" Length : {len(doc.page_content)} characters")
```

# Method 1 - CharacterTextSplitter

CharacterTextSplitter splits text into fixed-size chunks based on the number of characters, with optional overlap to preserve context between chunks.


```python
print("CharacterTextSplitter")


char_splitter = CharacterTextSplitter(
    separator="\n", #split on new lines
    chunk_size = 200 , #max chunk size in character
    chunk_overlap = 20 ,#overlap between chunks
    length_function=len #how to measure chunk size
)


chunks = char_splitter.split_documents(documents)


# Loop through each chunk and also get its number (starting from 1)
for i, chunk in enumerate(chunks, start=1):

    # Print the chunk number
    print(f"Chunk {i}")

    # Print a separator line for better readability
    print("-" * 40)

    # Print the actual text stored in the chunk
    print(chunk.page_content)

    # Print a blank line between chunks
    print()
```

# Method 2 - RecursiveCharacterTextSplitter

RecursiveCharacterTextSplitter splits text by trying multiple separators, preserving sentence and paragraph boundaries before falling back to characters.

```python

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,          # Maximum characters in each chunk
    chunk_overlap=20,        # Overlap between consecutive chunks
    separators=["\n\n", "\n", " ", ""],  # Order of separators to try
    length_function = len
)

# Split the document into chunks
chunks = text_splitter.split_documents(documents)

# Display the chunks
for i, chunk in enumerate(chunks, start=1):

    # Print the chunk number
    print(f"Chunk {i}")

    print(f"Length: {len(chunk.page_content)} characters")


    # Print a separator line
    print("-" * 40)



    # Print the chunk content
    print(chunk.page_content)

    # Blank line for readability
    print()
```

# Method 3 - TokenTextSplitter

TokenTextSplitter splits text by tokens instead of characters, ensuring chunks fit within an LLM's context window without exceeding limits.

```python
# Create the token splitter
text_splitter = TokenTextSplitter(
    chunk_size=50,      # Maximum 50 tokens per chunk
    chunk_overlap=10    # Repeat 10 tokens in the next chunk
)

# Split the document into chunks
chunks = text_splitter.split_documents(documents)

# Display the chunks
for i, chunk in enumerate(chunks, start=1):

    # Print the chunk number
    print(f"Chunk {i}")

    # Print a separator line
    print("-" * 40)

    # Print the chunk content
    print(chunk.page_content)

    # Print a blank line
    print()
```
