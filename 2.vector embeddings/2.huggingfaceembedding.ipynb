# HuggingFace Embeddings

```python
from langchain_huggingface import HuggingFaceEmbeddings


#initialize simple embedding model 

embeddings = HuggingFaceEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2"
)

embeddings
```

```python
## create first embedding 

text = "hello i am learning rag to become ai engineer in 2026"

embeddding_vector = embeddings.embed_query(text)
print(f'text : {text}')
print(f"embeddings_len : {len(embeddding_vector)}")
print(embeddding_vector)
```
